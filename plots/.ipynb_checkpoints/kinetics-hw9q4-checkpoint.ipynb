```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Time (s)': [0, 120, 240, 360, 480, 600, 720, 840, 960, 1080],
    'Tracer concentration (g/m³)': [0.0, 6.5, 12.5, 12.5, 10.0, 5.0, 2.5, 1.0, 0.0, 0.0]
}

df = pd.DataFrame(data)
# make a t*E(t) column
df['tC(t)'] = df['Time (s)'] * df['Tracer concentration (g/m³)']

#plt.plot(df['Time (s)'],df['Tracer conc. (g/m³)'])
plt.plot(df['Time (s)'],df['tC(t)'])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

ac = np.trapz(df['Tracer concentration (g/m³)'], df['Time (s)'])

print(f"Area under the curve: {ac:.2f} g·s/m³")

# Optional: Visualize the curve and area
plt.figure(figsize=(7, 3))
plt.fill_between(df['Time (s)'], df['Tracer concentration (g/m³)'], alpha=0.2, color='skyblue')
plt.plot(df['Time (s)'], df['Tracer concentration (g/m³)'], 'bo-', linewidth=2)
plt.title('C(t)\n(Area = {:.2f} g·s/m³)'.format(ac))
plt.xlabel('Time (s)')
plt.ylabel('Tracer concentration (g/m³)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()


## tC(t)

atc = np.trapz(df['tC(t)'], df['Time (s)'])

print(f"Area under the curve: {atc:.2f} g·s/m³")

# Optional: Visualize the curve and area
plt.figure(figsize=(7,3))
plt.fill_between(df['Time (s)'], df['tC(t)'], alpha=0.2, color='purple')
plt.plot(df['Time (s)'], df['tC(t)'], 'ro-', linewidth=2)
plt.title('tC(t) \n(Area = {:.2f} g·s/m³)'.format(atc))
plt.xlabel('Time (s)')
plt.ylabel('tC(t)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
print(f"{atc/ac} is the mean residence time of this reactor.")
```
